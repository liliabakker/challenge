#Análises e tratamentos

##Dataset - Produto - Petróleo

```python
produto = ['Planilha1', 'Planilha2', 'Planilha3', 'Planilha4', 'Planilha5', 'Planilha6', 'Planilha7', 'Planilha8',\
           'Planilha9', 'Planilha10', 'Planilha12', 'Planilha13', 'Planilha14', 'Planilha15', 'Planilha16', 'Planilha17', \
           'Planilha18', 'Planilha19', 'Planilha20', 'Planilha21', 'Planilha22']
dados_produto = pd.DataFrame()
```

```python
for i in produto:
  tabela = pd.read_excel('/content/drive/MyDrive/Datasets/vendas-combustiveis-m3.xlsx', sheet_name= i)
  dados_produto = pd.concat([dados_produto, tabela], axis = 0, ignore_index = True)
  
```

```python
dados_produto.head()
```

```python
#Verificando inconsistencias na coluna "ANO"
dados_produto["ANO"].unique()
```

```python
#imprime o número de linhas e colunas do dataframe
dados_produto.shape
```

```python
df_produto = dados_produto.copy()
```

```python
#verificando se temos valores duplicados
df_produto.duplicated().sum()
```

```python
#Verificando informações do dataframe
df_produto.info()
```

```python
#verifica atributos com valores faltantes 
faltantes = [var for var in df_produto.columns if df_produto[var].isnull().sum() > 0]
```

```python
df_produto[faltantes].isnull().sum()
```

###Tratamento

```python
# Substituindo Nan por 0
df_produto.fillna(0, inplace= True)
```

```python
#verificando se os valores da tabela correspondem a tabela dinâmica
pd.set_option("display.precision", 3)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
tabela_1 = df_produto.groupby("ANO").sum().T
```

```python
tabela_1.head(15)
```

```python
#Drop de colunas
df_produto.drop(columns=["REGIÃO","TOTAL"], inplace=True)
```

```python
#Verificando itens coluna "ESTADO"
df_produto["ESTADO"].unique()
```

```python
#Renomeando atributos
df_produto.columns=["product", "ANO", "uf","unit", "01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
```

```python
#Renomeando itens atributo "uf"
(df_produto['uf'].replace({'RONDÔNIA':'RO', 'ACRE':'AC', 'AMAZONAS':'AM', 'RORAIMA':'RR', 'PARÁ':'PA', 'AMAPÁ':'AP',
       'TOCANTINS':'TO', 'MARANHÃO':'MA', 'PIAUÍ':'PI', 'CEARÁ':'CE', 'RIO GRANDE DO NORTE':'RN',
       'PARAÍBA':'PB', 'PERNAMBUCO':'PE', 'ALAGOAS':'AL', 'SERGIPE':'SE', 'BAHIA':'BA',
       'MINAS GERAIS':'MG', 'ESPÍRITO SANTO':'ES', 'RIO DE JANEIRO':'RJ', 'SÃO PAULO':'SP',
       'PARANÁ':'PR', 'SANTA CATARINA':'SC', 'RIO GRANDE DO SUL':'RS',
       'MATO GROSSO DO SUL':'MS', 'MATO GROSSO':'MT', 'GOIÁS':'GO', 'DISTRITO FEDERAL':'DF'}, inplace= True))
```

```python
df_produto.head()
```

```python
#manipualções para criação do atributo "year_month"
df_produto2 = df_produto.melt(id_vars=["product","uf","unit", "ANO"], value_vars=['01','02','03','04','05','06','07','08','09','10','11','12'],\
                var_name= "mês", value_name= "volume")
```

```python
df_produto2["year_month"] = df_produto2["ANO"].map(str) + "-" + df_produto2["mês"]
```

```python
df_produto2["year_month"] = pd.to_datetime(df_produto2["year_month"])
```

```python
df_produto2.insert(0, 'created_at', pd.to_datetime('now').replace(microsecond=0))
```

```python
#Drop de colunas
df_produto2.drop(columns=["ANO", "mês"], inplace=True)
```

```python
df_produto2 = df_produto2.reindex(columns=['year_month', 'uf', 'product', 'unit', 'volume', 'created_at'])
```

```python
df_produto2.head(5)
```

```python
#Verificando os tipos de dados finais
df_produto2.dtypes
```
